```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
hero=pd.read_csv('charcters_stats.csv')
```

```python
hero.head()
```

```python
hero.shape
```

```python
hero['Alignment'].value_counts()
```

```python
good=hero[hero['Alignment']=='good']
```

```python
good.head()
```

```python
good.sort_values(by=['Speed'],ascending=False).head(10)
```

```python
good.sort_values(by=['Total'],ascending=False).head(10)
```

```python
good.sort_values(by=['Power'],ascending=False).head(10)
```

```python
good_max_power=good[good['Power']==100]
```

```python
good_max_power.head(10)
```

```python
good_max_power.shape
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(good_max_power['Name'])[0:5],list(good_max_power['Total'])[0:5],color='r')
```

```python
plt.figure(figsize=(8,8))
plt.pie(list(good_max_power['Total'].value_counts()[0:5]),labels=list(good_max_power['Name'].value_counts()[0:5].keys()),autopct='%0.1f%%')
plt.show()
```
